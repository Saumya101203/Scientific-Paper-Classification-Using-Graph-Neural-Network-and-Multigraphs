```python
# Cell: Run ORIGINAL SAGE Benchmark on 20% Subset
# (Uses original experiments.py with adj_t and to_hetero)
#
import yaml
from pathlib import Path
import os

# --- Configuration ---
SUBSET_RATIO_TEST = 0.20 # Using 20%
MODEL_TO_RUN = 'SAGE'
DATASET_NAME = 'ogbnarxiv'

print(f"\n--- Running ORIGINAL {MODEL_TO_RUN} Benchmark on {SUBSET_RATIO_TEST*100:.0f}% Subset ---")

# --- Define the 20% subset file path ---
# Assume notebook is running from the project's root directory
project_root = Path.cwd()
os.chdir(project_root)
# Load base params to get paths if needed later
try:
    with open("config/experiments_config.yaml") as f:
        # Load base params but be careful not to modify the original dict globally
        base_params = yaml.load(f, Loader=yaml.FullLoader)
except FileNotFoundError:
    print("ERROR: Base config/experiments_config.yaml not found.")
    raise

subset_20_filename = f"{DATASET_NAME}_heterodata_subset_{int(SUBSET_RATIO_TEST*100)}pct.pt"
subset_20_path = project_root / 'data' / subset_20_filename
if not subset_20_path.exists():
    raise FileNotFoundError(f"20% subset file not found at: {subset_20_path}. Please create it first.")
else:
    print(f"Using 20% subset file: {subset_20_path}")


# --- Create experiment config specifically for this run ---
# Create a fresh dictionary to avoid modifying the global 'params'
exp_params = {
    'dataset': DATASET_NAME,
    'data': {'graph_dataset': {DATASET_NAME: str(subset_20_path)}},
    'node_embs': 'preloaded_in_subset', # Assume features are in the subset file
    # Use all four edge types (as expected by the original script's utils function)
    'edge_type_selection': {DATASET_NAME: ['references', 'author', 'venue', 'fos']},
    'optimizer': {'lr': 0.001, 'weight_decay': 0},
    'model': {'name': MODEL_TO_RUN, 'hidden_channels': 128, 'num_layers': 2, 'dropout': 0.5},
    'early_stop_threshold': 30,
    'epochs': 500, # Use original epoch count
    'runs': 3 # Run 3 times for average
}

# --- Save config and run ---
temp_config_path = f"config/temp_{MODEL_TO_RUN.lower()}_original_20pct_config.yml"
with open(temp_config_path, 'w') as f: yaml.dump(exp_params, f, sort_keys=False)

print(f"\nTemporary config saved to: {temp_config_path}")
print("--- Config Contents ---")
with open(temp_config_path, 'r') as f: print(f.read())
print("-----------------------")

print(f"\nRunning training for {MODEL_TO_RUN} (Original script, 20% data)...")
# Run WITHOUT generate_preds_path as we only need the accuracy score
# Ensure you are running the ORIGINAL experiments.py
%run scripts/experiments.py --config {temp_config_path}

print(f"\nStandalone {MODEL_TO_RUN} benchmark finished.")
```

```python
import pandas as pd

METADATA_PATH = "data/tables/ogbnarxiv_mag_metadata.parquet.gzip"
print(f"--- Loading Metadata from {METADATA_PATH} ---")
df = pd.read_parquet(METADATA_PATH)

print("\n--- All Columns and Example Data ---")
for col in df.columns:
    print(f"\nColumn: '{col}'")
    # Print the first 5 unique, non-null values
    try:
        unique_values = df[col].dropna().unique()
        print(f"  Total Unique Values: {len(unique_values)}")
        print("  Example Values:")
        for val in unique_values[:5]:
            print(f"    - {val}")
    except Exception as e:
        print(f"  Could not get unique values (unhashable?): {e}")
        print(f"  First 5 values: \n{df[col].head()}")

print("\n--- Verdict ---")
print("Look for a column that is *not* 'venue' but looks like a conference or journal name.")
```
