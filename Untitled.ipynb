```python
#
# Cell: Generate 20% Subset
#
import os
from pathlib import Path

# --- Configuration ---
FULL_DATASET_NAME = 'ogbnarxiv_heterodata.pt'
SUBSET_RATIO_20 = 0.20  # Using 20%
DATASET_NAME = 'ogbnarxiv'

project_root = Path.cwd()

# Define the input and output paths
FULL_DATA_PATH = project_root / 'data' / FULL_DATASET_NAME
SUBSET_20_FILENAME = f"{DATASET_NAME}_heterodata_subset_{int(SUBSET_RATIO_20*100)}pct.pt"
SUBSET_20_PATH = project_root / 'data' / SUBSET_20_FILENAME

if not FULL_DATA_PATH.exists():
    raise FileNotFoundError(f"Full dataset not found at: {FULL_DATA_PATH}. Please check the file name.")

print(f"\n--- Generating {SUBSET_RATIO_20*100:.0f}% Subset ---")

# Execute the subset creation script
# Make sure your corrected create_subset.py is saved before running this.
%run scripts/create_subset.py --input_file "{FULL_DATA_PATH}" --output_file "{SUBSET_20_PATH}" --ratio {SUBSET_RATIO_20}

print(f"\n✅ 20% subset file is ready at: {SUBSET_20_PATH}")
```
