```python
#
# Cell: Run RevGAT Benchmark on 20% Subset
#
import yaml
from pathlib import Path
import os

# --- Configuration ---
SUBSET_RATIO_TEST = 0.20
MODEL_TO_RUN = 'RevGAT' # Changed from SAGE to RevGAT
DATASET_NAME = 'ogbnarxiv'

print(f"\n--- Running {MODEL_TO_RUN} Benchmark on {SUBSET_RATIO_TEST*100:.0f}% Subset ---")

# --- Define the 20% subset file path ---
project_root = Path.cwd()
os.chdir(project_root)

subset_20_filename = f"{DATASET_NAME}_heterodata_subset_{int(SUBSET_RATIO_TEST*100)}pct.pt"
subset_20_path = project_root / 'data' / subset_20_filename
if not subset_20_path.exists():
    raise FileNotFoundError(f"20% subset file not found at: {subset_20_path}. Please create it first.")
else:
    print(f"Using 20% subset file: {subset_20_path}")


# --- Create experiment config specifically for this run ---
exp_params = {
    'dataset': DATASET_NAME,
    'data': {'graph_dataset': {DATASET_NAME: str(subset_20_path)}},
    'node_embs': 'preloaded_in_subset',
    'edge_type_selection': {DATASET_NAME: ['references', 'author', 'venue', 'fos']},
    'optimizer': {'lr': 0.001, 'weight_decay': 0},
    'model': {
        'name': MODEL_TO_RUN,          # Set to RevGAT
        'hidden_channels': 128,        # Total hidden dimension
        'num_layers': 2,               # Number of reversible blocks
        'dropout': 0.5,
        'heads': 4                     # Added 'heads' parameter for attention
    },
    'early_stop_threshold': 30,
    'epochs': 500,
    'runs': 3
}

# --- Save config and run ---
temp_config_path = f"config/temp_{MODEL_TO_RUN.lower()}_original_20pct_config.yml"
with open(temp_config_path, 'w') as f:
    yaml.dump(exp_params, f, sort_keys=False)

print(f"\nTemporary config saved to: {temp_config_path}")
print("--- Config Contents ---")
with open(temp_config_path, 'r') as f:
    print(f.read())
print("-----------------------")

print(f"\nRunning training for {MODEL_TO_RUN} (Original script, 20% data)...")
# Execute your updated experiments.py script with the new config
%run scripts/experiments.py --config {temp_config_path}

print(f"\nStandalone {MODEL_TO_RUN} benchmark finished.")
```

```python
#
# Cell: Run RevGAT Benchmark on 40% Subset
#
import yaml
from pathlib import Path
import os

# --- Configuration ---
SUBSET_RATIO_TEST = 0.40  # Using 40% of the dataset
MODEL_TO_RUN = 'RevGAT'   # Reversible GAT model
DATASET_NAME = 'ogbnarxiv'

print(f"\n--- Running {MODEL_TO_RUN} Benchmark on {SUBSET_RATIO_TEST*100:.0f}% Subset ---")

# --- Define the 40% subset file path ---
project_root = Path.cwd()
os.chdir(project_root)

subset_40_filename = f"{DATASET_NAME}_heterodata_subset_{int(SUBSET_RATIO_TEST*100)}pct.pt"
subset_40_path = project_root / 'data' / subset_40_filename

if not subset_40_path.exists():
    raise FileNotFoundError(f"40% subset file not found at: {subset_40_path}. Please create it first.")
else:
    print(f"Using 40% subset file: {subset_40_path}")


# --- Create experiment config specifically for this run ---
exp_params = {
    'dataset': DATASET_NAME,
    # POINTING TO THE NEW 40% SUBSET PATH
    'data': {'graph_dataset': {DATASET_NAME: str(subset_40_path)}},
    'node_embs': 'preloaded_in_subset',
    'edge_type_selection': {DATASET_NAME: ['references', 'author', 'venue', 'fos']},
    'optimizer': {'lr': 0.001, 'weight_decay': 0},
    'model': {
        'name': MODEL_TO_RUN,
        'hidden_channels': 64,    # Hidden dimension per stream
        'num_layers': 2,           # Number of reversible blocks
        'dropout': 0.5,
        'heads': 2                 # Multi-head attention parameter
    },
    'early_stop_threshold': 30,
    'epochs': 500,
    'runs': 3  # Run multiple times for averaging
}

# --- Save config and run ---
temp_config_path = f"config/temp_{MODEL_TO_RUN.lower()}_original_40pct_config.yml"
with open(temp_config_path, 'w') as f:
    yaml.dump(exp_params, f, sort_keys=False)

print(f"\nTemporary config saved to: {temp_config_path}")
print("--- Config Contents ---")
with open(temp_config_path, 'r') as f:
    print(f.read())
print("-----------------------")

print(f"\nRunning training for {MODEL_TO_RUN} (Original script, 40% data)...")
# Execute your updated experiments.py script with the new config
%run scripts/experiments.py --config {temp_config_path}

print(f"\nStandalone {MODEL_TO_RUN} benchmark on 40% finished.")
```
