```python
import torch

# Force MPS if available
if torch.backends.mps.is_available():
    device = torch.device("mps")
    print("✅ Using MPS backend")
elif torch.cuda.is_available():
    device = torch.device("cuda")
    print("✅ Using CUDA backend")
else:
    device = torch.device("cpu")
    print("⚠️ Falling back to CPU")

# Monkey-patch torch.device globally if experiments.py uses 'cuda'/'cpu'
torch_device_orig = torch.device
def torch_device_patched(dev, *args, **kwargs):
    if dev in ["mps", "cpu"]:
        return device
    return torch_device_orig(dev, *args, **kwargs)
torch.device = torch_device_patched
```
